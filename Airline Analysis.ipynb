# Analytics
## Coach price distribution
As mentioned above, the first question to address is, "What is the average coach price, and is 500 `dolar` a reasonable price for a coach ticket?" To answer this, we will first create a box plot.

Based on the box plot (figure 1), the coach ticket prices range from 100 dolar to 500 `dolar`. The highest outlier is 600 `dolar`, while the lowest outlier is 0 `dolar`. The first quartile (Q1) is 190 `dolar`, and the average coach price is 376 `dolar`. Therefore, 500 `dolar` seems a bit high for a coach ticket.

To have a clearly look regard to coach price distribution we look at figure 2, it show that the histogram of coach price is standard and the center of the distribute from 350 `dolar` to 400 `dolar`

```python
import seaborn as sns
import matplotlib.pyplot as plt
## Task 1
print('Coach price mean: $' + str(np.mean(flight.coach_price)))
print('Coach price median: $' + str(np.median(flight.coach_price)))
sns.boxplot(flight.coach_price)
plt.xlabel('Coach price')
plt.title('Figure 1: Boxplot coach price')
plt.savefig('figure1.png')
plt.show()
plt.clf()
sns.histplot(flight.coach_price)
plt.title('Figure 2: Histogram of coach price')
plt.savefig('figure2.png')
plt.show()
plt.clf()
```

## Coach Ticket Prices for 8-Hour Flights

In this section, we focus on visualizing the coach ticket prices for flights that are 8 hours long. By examining this subset of data, we aim to understand the price distribution and determine if a \$500 ticket is more reasonable for such flights compared to the general price range observed earlier.

## Visualization and Analysis

To analyze the coach ticket prices for 8-hour flights, a box plot and histogram were generated. These visualizations reveal key insights into the distribution of prices.

- **High Price:** The highest recorded coach ticket price for 8-hour flights is \$580.
- **Low Price:** The lowest coach ticket price observed is \$120.
- **Average Price:** The average ticket price for 8-hour flights is approximately \$410.

Comparing this with the earlier analysis of general coach ticket prices, we can observe that the average price for 8-hour flights is higher than the overall average coach price (\$376). This suggests that longer flights, such as those lasting 8 hours, tend to command higher ticket prices.

## Reasonableness of a \$500 Ticket

Given the higher average and maximum prices for 8-hour flights, a \$500 ticket appears to be more reasonable in this context than it did when considering all flight durations. While it still falls on the higher end of the price range, it is less of an outlier and may be justified by the longer flight duration and associated costs.

```python
## Task 2
# Assuming flight is your DataFrame and is already loaded
print(np.mean(flight.coach_price[flight.hours == 8]))
print(np.median(flight.coach_price[flight.hours == 8]))

coach_price_8hr_flight = flight.coach_price[flight.hours == 8]

# Create the boxplot
ax = plt.figure(figsize = (10,8))
ax = plt.subplot()
sns.boxplot(y=coach_price_8hr_flight)
plt.xlabel('8 hours flight')
plt.title('8 Hours flight coach price')
plt.savefig('figure3.png')
plt.show()
plt.clf()
```

## Distribution of Flight Delay Times

In this section, we explore the distribution of flight delay times to better understand the typical delays that passengers may experience. This analysis is particularly important for planning connecting flights, where short layovers could result in missed connections if delays are frequent or substantial.
#### Visualization and Analysis

The histogram in Figure 3 illustrates the distribution of delay times across all flights, with delays capped at 50 minutes for clarity. From the distribution, several key insights emerge:

- **Common Delay Times:** The majority of delays fall between 5 and 15 minutes, with a significant peak around the 10-minute mark. This suggests that short delays are common and should be factored into planning connections.
- **Significant Delays:** While less frequent, there are delays that extend up to 50 minutes, indicating that longer delays do occur, albeit with lower frequey.

#### Implications for Connecting Flights

Given the distribution of delays, it is evident that most flights experience relatively short delays. However, the presence of longer delays, though less common, means that connecting flights should ideally be scheduled with a buffer to account for these potential disruptions. For clients with tight connections, it would be advisable to allow extra time between flights to reduce the risk of missed connections due to unexpected delays.

In summary, while short delays are typical, the variability in delay times suggests that cautious scheduling of connecting flights is essential to ensure a smooth travel experience for ients.

```python
## Task 3
sns.histplot(flight.delay[flight.delay <= 500], bins = 50)
plt.xlabel('Delay minutes')
plt.title('Number of delay time')
plt.savefig('figure4.png')
plt.show()
plt.clf()
```

## Relationship Between Coach and First-Class Prices

This section examines the correlation between coach and first-class ticket prices. The scatter plot with a trend line in Figure 4 shows a positive relationship between the two: as coach prices increase, first-class prices tend to rise as well.

While higher coach prices generally correspond with higher first-class prices, there is some variability, indicating that other factors may also influence first-class pricing.

In summary, coach and first-class prices are positively correlated, but the relationship isn't perfectly linear.

```python
## Task 4
perc = 0.1
flight_sub = flight.sample(n = int(flight.shape[0]*perc))
sns.lmplot(x = "coach_price", y = "firstclass_price", data = flight_sub, line_kws={'color': 'black'}, lowess=True)
plt.savefig('figure5.png')
plt.show()
plt.clf()
```

## Relationship Between Coach Prices and Inflight Features

This section examines how inflight features—meals, entertainment, and WiFi—affect coach ticket prices.
### Inflight Meals

The histogram indicates that flights with meals have slightly higher prices. However, the distribution is similar to flights without meals, despite more data for the latter

#### Inflight Entertainment

Flights with entertainment are generally more expensive, as shown by a shift towards higher prices in the distribuon.

#### Inflight WiFi

Flights with WiFi access are also priced higher, reflecting the added value of internet conetivity.

Inflight meals do not significantly affect coach prices, while entertainment and WiFi are linked to higher ticket costs.

```python
## Task 5
# Inflight Meals
sns.histplot(flight, x = "coach_price", hue = flight.inflight_meal)
plt.title('Coach price of inflight meal')
plt.savefig('figure6.png')
plt.show()
plt.clf()

# Inflight Entertainment
sns.histplot(flight, x = "coach_price", hue = flight.inflight_entertainment)
plt.title('Coach price of inflight entertaiment')
plt.savefig('figure7.png')
plt.show()
plt.clf()

# Inflight WiFi
sns.histplot(flight, x = "coach_price", hue = flight.inflight_wifi)
plt.title('Coach price of inflight wifi')
plt.savefig('figure8.png')
plt.show()
plt.clf()
```

## Relationship between number of passengers and flight hours 

Below is the scatter plot show the relationship between number of passengers and hours flight:

- Short Flights (1 to 5 hours): There is a dense concentration of data points, indicating less variation in the number of passengers. This suggests that flights lasting between 1 to 5 hours tend to have a relatively consistent passenger count. Factors such as flight schedule, convenience, and pricing may contribute to this stability.- 
Long Flights (6 to 8 hours): In contrast, there are fewer data points for flights lasting between 6 to 8 hours. This implies that longer flights have fewer passengers overall. It’s possible that passengers prefer shorter flights due to comfort, convenience, or other factors.

```python
## Task 6
sns.lmplot(x = "hours", y = "passengers", data = flight_sub, x_jitter = 0.25, scatter_kws={"s": 5, "alpha":0.2}, fit_reg = False)
plt.title('Hours of flight with number of passengers')
plt.savefig('figure9.png')
plt.show()
plt.clf()
```

## First class price different between weekend

- Price Correlation: There is a positive correlation between coach prices and first-class prices, meaning that as the price of coach tickets increases, the price of first-class tickets also tends to increase.- 
Weekend vs  Weekday Prices: The data points are color-coded to distinguish between weekends (orange) and weekdays (blue). The graph shows that both coach and first-class prices are generally higher on weekends compared to weekdays. This is evident from the higher positioning of the orange data points on both axes.

This visualization highlights how airlines might adjust their pricing strategies based on the day of travel, with higher prices typically seen on weekends. This could be due to increased demand for travel during weekends.

```python
## Task 7
perc = 0.1
flight_sub = flight.sample(n = int(flight.shape[0]*perc))
sns.scatterplot(x = "coach_price", y = "firstclass_price", hue = 'weekend', data = flight_sub, alpha = 0.55)
plt.title('Coach price and first class price in the weekend compare to weekdays')
plt.savefig('figure10.png')
plt.show()
plt.clf()
```
